### Tree and Random Forest

````r
library(rpart)
install.packages("rpart.plot")
install.packages("rattle")
library(rpart.plot)  
library(rattle)
install.packages("randomForest")
library(randomForest)
````

````r
prop = table(train$Revenue) / dim(train)[1]
w = rep(1/prop[2], dim(train)[1])
w[which(train$Revenue == FALSE)] = 1/prop[1]

tree1 = rpart(as.factor(Revenue) ~ ., train, parms = list(split = "gini"), weights = w)
plotcp(tree1)
````

````r
tree.gini = rpart(as.factor(Revenue) ~ ., train, parms = list(split = "gini"), method = "class", cp = 0.083)
fancyRpartPlot(tree.gini)
````

````r
# results from single tree

train.pred=predict(tree.gini, train, type="class")
table(train.pred, train$Revenue)
sum(train.pred != train$Revenue) / dim(train)[1]
test.pred=predict(tree.gini, test, type="class")
table(test.pred, test$Revenue)
sum(test.pred != test$Revenue) / dim(test)[1]
````

````r
# do some exploration for the optimal number of predictors to use

train_x <- train %>% select(-Revenue)
test_x <- test %>% select(-Revenue)
test_errors <- c()
train_errors <- c()
mtrys <- seq(2,ncol(train_x), by = 1)

for (i in mtrys) {
  rf.rev <- randomForest(as.factor(Revenue)~.,mtry = i,ntree = 1000,importance = TRUE,data = train)
  train_pred <- predict(rf.rev, train_x, type="class")
  test_pred <- predict(rf.rev, test_x, type="class")
  train_error <- mean(train_pred != train$Revenue)
  train_errors <- c(train_errors, train_error)
  test_error <- mean(test_pred != test$Revenue)
  test_errors <- c(test_errors, test_error)
}

ggplot() + geom_line(mapping = aes(x = mtrys, y = train_errors, color = "training")) + 
  geom_line(mapping = aes(x = mtrys, y = test_errors, color = "testing")) + 
  labs(title = "Error Rates v Number of Predictors", x = "Number of Predictors", 
       y = "Error") + theme_bw()
````

````r
# do some exploration for the optimal node size

nodesizes <- seq(1, 100, by = 5)
train_x <- train %>% select(-Revenue)
test_x <- test %>% select(-Revenue)
test_errors <- c()
train_errors <- c()

for (i in nodesizes) {
  rf.rev.node <- randomForest(as.factor(Revenue) ~ ., mtry = 4, ntree = 1000, nodesize = i, 
                          importance = TRUE, data = train)
  train_pred <- predict(rf.rev.node, train_x, type="class")
  test_pred <- predict(rf.rev.node, test_x, type="class")
  train_error <- mean(train_pred != train$Revenue)
  train_errors <- c(train_errors, train_error)
  test_error <- mean(test_pred != test$Revenue)
  test_errors <- c(test_errors, test_error)
}

ggplot() + geom_line(mapping = aes(x = nodesizes, y = train_errors, color = "training")) + 
  geom_line(mapping = aes(x = nodesizes, y = test_errors, color = "testing")) + 
  labs(title = "Error Rates v Nodesize", x = "Nodesize", y = "Error") + theme_bw()
````

````r
# optimal based on the above

rf_revenue <- randomForest(as.factor(Revenue) ~ ., mtry = 7, ntree = 1000, nodesize = 65, importance = TRUE, data = train)
train_x <- train %>% select(-Revenue)
test_x <- test %>% select(-Revenue)
train_pred <- predict(rf_revenue, train_x)
test_pred <- predict(rf_revenue, test_x)
````

## SVM

SVM only uses numerical, no categorical variables, so have to remove those first

````r
# install.packages("e1071")
library(e1071)
````

````r
numerical_train <- train %>% 
  dplyr::select(Administrative, Administrative_Duration, Informational, Informational_Duration,
                ProductRelated, ProductRelated_Duration, BounceRates, ExitRates, PageValues, Revenue)

numerical_test <- test %>% 
  dplyr::select(Administrative, Administrative_Duration, Informational, Informational_Duration,
                ProductRelated, ProductRelated_Duration, BounceRates, ExitRates, PageValues, Revenue)
````

````r
# performing some validation with linear kernel

tune.svm=tune(svm,as.factor(Revenue)~.,data=numerical_train,kernel="linear", ranges=list(cost=c(0.001, 0.01, 0.1, 1,5,10,100)))
summary(tune.svm)

best <- tune.svm$best.model
summary(best)
````

````r
# let's see how the model performed

# train set
ypred_train<- predict(best ,numerical_train)
table(predict=ypred_train, truth=numerical_train$Revenue)
# test set
ypred_test <- predict(best , numerical_test)
table(predict=ypred_test, truth= numerical_test$Revenue)
````

````r
# let's try a radial kernel

tune.radial=tune(svm, as.factor(Revenue)~., data=numerical_train, kernel="radial", ranges=list(cost=c(0.1,1,10,100,1000),gamma=c(0.5,1,2,3,4)))
summary(tune.radial)
````

````r
# how did radial do?
# train set
table(true=numerical_train$Revenue, pred=predict(tune.radial$best.model, newdata=numerical_train))
# test set
table(true=numerical_test$Revenue, pred=predict(tune.radial$best.model, newdata=numerical_test))
````
